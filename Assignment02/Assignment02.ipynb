#Data

\hat{y} = 4x + 50\\\\
y = \hat{y} + n, n\ is\ N(0,\sigma^2) \\\\$

### Domain of the functions
$ -100\ {\leqq}\ X\ {\leqq}\ 100$

```
m = 1000
x = np.arange(-100, 100, 0.1)
y_ = 4*x + 50
n = np.int64(np.random.normal(loc=0, scale=100, size=m))
m_list = np.int64(np.random.normal(loc=0, scale=30, size=m))
y = 4*m_list+50
data = y + n
```

```
# Testing code
plt.figure(figsize=(10,6))
plt.plot(x, y_, color='blue')
plt.plot(m_list, data, 'ro', color='black')
plt.grid(axis='both', which='major')
```

# Gradient Descent

{\theta}_{1}^{(t+1)} := {\theta}_{1}^{(t)} - {\alpha} \frac{1}{m}{\Sigma}_{i=1}^{m​}(h_{\theta}(x^{(i)})−y^{(i)})x^{(i)}\\\\
